```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#4-2-1 generator function that generates 10 integers and use it to build an array

def generate():
    for i in range(10):
        yield i
array = np.fromiter(generate(), dtype=float)
print(array)
```

```python
#4.2.2 two random array A and B, check if they are equal

A = np.random.randint(0, 10, 5)
B = np.random.randint(0, 10, 5)
equal = np.allclose(A, B)
print(A)
print(B)
if(equal):
    print("They are equal")
else:
    print("Not equal") 
```

```python
#4.2.3 random vector with shape (100,2) representing coordinates, find point by point distances

vector = np.random.random((100, 2))
x, y = np.atleast_2d(vector[:,0], vector[:,1])
distance = np.sqrt(pow((x-x.T),2)+pow((y-y.T), 2))
print(distance)
```

```python
#4.2.4 Subtract the mean of each row of a matrix

matrix = np.random.rand(3, 4)
sub = matrix - matrix.mean(axis=1, keepdims=True)
print(matrix)
print(sub)
```

```python
#4.2.5 sort an array by the nth column

new_arr = np.random.randint(0, 10, (5,5))
n = 2
sorted = new_arr[new_arr[:,n].argsort()]
print(new_arr)
print(sorted)
```

```python
#4.2.6 Compute a matrix rank

new_matrix = np.random.randint(0, 10, (5,5))
rank = np.linalg.matrix_rank(new_matrix)
print(new_matrix)
print("rank of the matrix:", rank)
```

```python
#4.2.6 get the block-sum (block size is 4x4)

arr = np.random.random((16,16))
k = 4
sum = np.add.reduceat(np.add.reduceat(arr, np.arange(0, arr.shape[0], k), axis=0),np.arange(0, arr.shape[1], k), axis=1)
print(arr)
print("sum:", sum)
```
